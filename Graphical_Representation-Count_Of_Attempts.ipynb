# Graphical representation 

```python
import matplotlib.pyplot as plt

def visualize_tasks():
    tasks = []
    dates = []
    scores = []
    attempts = 0
    
    while True:
        task = input("Enter the task name (or 'exit' to finish): ")
        
        if task.lower() == 'exit':
            break
        
        date = input("Enter the submission date (YYYY-MM-DD): ")
        score = int(input("Enter the score (out of 5): "))
        attempts += 1
        
        tasks.append(task)
        dates.append(date)
        scores.append(score)
        
        if score == 5:
            print(f"Congratulations! You achieved a perfect score on attempt {attempts} on {date}.\n")
    
    plt.figure(figsize=(10,5))
    plt.plot(range(1, attempts + 1), scores,marker='o')
    plt.xticks(range(1, attempts + 1), dates, rotation=45)
    plt.xlabel("Submission Date")
    plt.ylabel("Score (out of 5)")
    plt.title("User Task Scores")
    plt.show()

if __name__ == "__main__":
    visualize_tasks()
```
